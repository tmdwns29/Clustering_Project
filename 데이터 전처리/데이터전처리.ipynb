```python
import folium
import geopandas as gpd
import numpy as np
import pandas as pd
from shapely.geometry import box
from folium.plugins import HeatMap

# 1. GeoJSON 데이터 로드
gdf = gpd.read_file('/content/hangjeongdong_대전광역시.geojson')

# 2. 지도 범위 설정
minx, miny, maxx, maxy = gdf.total_bounds

# 3. 100x100 격자 생성
grid_cells = []
x_coords = np.linspace(minx, maxx, 101)
y_coords = np.linspace(miny, maxy, 101)

for i in range(len(x_coords) - 1):
    for j in range(len(y_coords) - 1):
        grid_cells.append(box(x_coords[i], y_coords[j], x_coords[i+1], y_coords[j+1]))

grid = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs=gdf.crs)

# 4. 격자와 행정구역 교차 계산
intersections = gpd.overlay(grid, gdf, how='intersection', keep_geom_type=False)

# 5. 각 격자에 대해 최대 면적을 가지는 행정구역 선택
intersections['area'] = intersections.area
largest_intersections = intersections.loc[intersections.groupby(intersections.index)['area'].idxmax()]

# 6. 쓰레기통 위치 데이터 로드
trash_bins = pd.read_csv('/content/유성구_공공쓰레기통.csv', encoding = 'cp949')
trash_bins_gdf = gpd.GeoDataFrame(trash_bins,
                                  geometry=gpd.points_from_xy(trash_bins.경도, trash_bins.위도),
                                  crs="EPSG:4326")

# 7. Folium 지도를 생성하고 격자 및 쓰레기통 위치 표시
m = folium.Map(location=[36.5, 127.5], zoom_start=12, tiles='cartodbpositron')

# 격자 데이터를 GeoJSON으로 변환하여 지도에 추가
folium.GeoJson(
    largest_intersections.to_json(),
    name='grid',
    style_function=lambda x: {'color': 'blue', 'weight': 1},
    tooltip=folium.GeoJsonTooltip(fields=['adm_nm'], aliases=['행정구역'], localize=True)
).add_to(m)

# 쓰레기통 위치를 HeatMap으로 시각화
heat_data = [[row['위도'], row['경도']] for index, row in trash_bins.iterrows()]
HeatMap(heat_data).add_to(m)

# 레이어 제어 추가
folium.LayerControl().add_to(m)

# 지도 저장 및 표시
m.save('yuseong_trashbin_density_map.html')
m
```
