```python
import pandas as pd
import glob

file_paths = glob.glob('/content/drive/MyDrive/유성구 데이터/유성구_법정동별 소비금액 데이터*.csv')
file_paths

df = []
for path in file_paths:
  try:
    df.append(pd.read_csv(path, encoding='CP949'))
  except UnicodeDecodeError:
    df.append(pd.read_csv(path))
```

```python
# 데이터프레임 합치기
merged_df = pd.concat(df, ignore_index=True)

# '숙박 및 음식점업(55-56)' 필터링
filtered_df = merged_df[merged_df['표준산업대분류명'] == '숙박 및 음식점업(55-56)']

# 법정동코드, 법정동명 별로 이용금액 합계 계산
result_df = filtered_df.groupby(['법정동코드', '법정동명'])['이용금액'].sum().reset_index()

result_df.to_csv('/content/drive/MyDrive/유성구 데이터/법정동별 숙박 및 음식점업 소비금액_데이터.csv', index=False)
```

```python
merged_df = pd.concat(df, ignore_index=True)

result_df = merged_df[merged_df['표준산업대분류명'] == '숙박 및 음식점업(55-56)'].groupby('법정동명')['이용금액'].sum().reset_index()
result_df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
result_df.to_csv('/content/법정동별 숙박 및 음식점업 소비금액 데이터_2022.csv', index=False)
```

```python
# 새로운 데이터 추가
new_data = {'법정동명': ['금탄동'], '이용금액': [0]}  # 예시 금액
new_df = pd.DataFrame(new_data)

# 기존 데이터프레임에 새로운 데이터 추가
result_df = pd.concat([result_df, new_df], ignore_index=True)
```
