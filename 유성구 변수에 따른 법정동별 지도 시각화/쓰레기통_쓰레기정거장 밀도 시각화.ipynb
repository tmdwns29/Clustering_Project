```python
import geopandas as gpd
import folium

# 1. Shapefile 로드
gdf = gpd.read_file('/content/TL_SCCO_EMD.shp', encoding='euc-kr')  # 실제 shapefile 경로로 수정하세요

# 2. CRS 설정 및 변환
gdf = gdf.set_crs(epsg=5179)  # 원본 CRS 설정
gdf = gdf.to_crs(epsg=4326)   # EPSG:4326으로 변환

# 3. GeoDataFrame을 GeoJSON으로 변환
geojson_data = gdf.to_json()

# 4. 부산 중심부의 경도와 위도 (위도와 경도를 수정하여 원하는 중심으로 설정하세요)
center = [36.3668, 127.3451]  # 위도와 경도

# 기본 지도 생성
m = folium.Map(location=center, zoom_start=10)

# GeoJSON 레이어 추가
folium.GeoJson(
    geojson_data,
    name='geojson'
).add_to(m)

# 레이어 컨트롤 추가
folium.LayerControl().add_to(m)

# 지도 저장
m.save('shapefile_map.html')

# 지도 출력
m
```

```python
import folium
from folium import Choropleth, LayerControl
import geopandas as gpd

# 부산 중심부의 경도와 위도
center = [36.3668, 127.3451]

# 기본 지도 생성
m = folium.Map(location=center, zoom_start=10, tiles='CartoDB positron')

# GeoDataFrame을 GeoJSON으로 변환
geojson_data = gdf.to_json()  # gdf는 법정동 경계를 가진 GeoDataFrame

# Choropleth 레이어 추가
Choropleth(
    geo_data=geojson_data,
    data=trash,  # m_df는 법정동명과 예상쓰레기발생량이 담긴 데이터프레임
    columns=['법정동명','사용유무'],  # '법정동명'과 '예상쓰레기발생량'이 들어있는 컬럼
    key_on='feature.properties.법정동명',  # GeoJSON의 속성 키
    fill_color='Greens',  # 색상 팔레트 (Yellow to Red)
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='쓰레기통 수 밀도'
).add_to(m)

# 법정동별 이름 추가 (텍스트로 표시)
for _, row in gdf.iterrows():
    # 법정동의 중심 좌표 계산 (Polygon의 centroid)
    centroid = row['geometry'].centroid
    lat, lon = centroid.y, centroid.x

    # 텍스트를 DivIcon으로 추가
    folium.Marker(
        location=[lat, lon],
        icon=folium.DivIcon(
            html=f'<div style="font-size: 8pt; font-weight: bold; color: black; transform: rotate(0deg); text-align: center; white-space: nowrap;">{row["법정동명"]}</div>',
            # 텍스트의 크기를 8pt로 줄이고, 회전은 0도로 가로로 표시
        ),
        tooltip=row['법정동명']
    ).add_to(m)

# 레이어 컨트롤 추가
LayerControl().add_to(m)

# 지도 저장 및 출력
m.save('yeongcheon_map.html')
m
```

```python
import folium
from folium import Choropleth, LayerControl

# 부산 중심부의 경도와 위도
center = [36.3668, 127.3451]

# 기본 지도 생성
m = folium.Map(location=center, zoom_start=10, tiles='CartoDB positron')

# GeoDataFrame을 GeoJSON으로 변환
geojson_data = gdf.to_json()

# Choropleth 레이어 추가
Choropleth(
    geo_data=geojson_data,
    data=gdf,  # gdf는 법정동명과 쓰레기통수가 담긴 데이터프레임
    columns=['법정동명', '쓰레기통수'],  # 법정동명과 쓰레기통 수
    key_on='feature.properties.법정동명',  # GeoJSON의 속성 키
    fill_color='Greens',  # 초록색 계열로 밀도 시각화
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='쓰레기통 수 밀도'
).add_to(m)

# 법정동별 이름 추가 (텍스트로 표시)
for _, row in gdf.iterrows():
    # 법정동의 중심 좌표 계산 (Polygon의 centroid)
    centroid = row['geometry'].centroid
    lat, lon = centroid.y, centroid.x

    # 텍스트를 DivIcon으로 추가
    folium.Marker(
        location=[lat, lon],
        icon=folium.DivIcon(
            html=f'<div style="font-size: 8pt; font-weight: bold; color: black; transform: rotate(0deg); text-align: center; white-space: nowrap;">{row["법정동명"]}</div>',
            # 텍스트의 크기를 8pt로 줄이고, 회전은 0도로 가로로 표시
        ),
        tooltip=row['법정동명']
    ).add_to(m)

# 레이어 컨트롤 추가
LayerControl().add_to(m)

# 지도 저장 및 출력
m.save('trashcounts.jpg')
m
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# GeoDataFrame을 로드합니다.
fp = "/content/TL_SCCO_EMD.shp"  # shapefile 경로를 지정하세요.
gdf = gpd.read_file(fp)

# 데이터 프레임 준비 (쓰레기통 수를 포함한 GeoDataFrame)
# 예를 들어, gdf에는 '법정동명'과 '쓰레기통수' 컬럼이 있어야 합니다.

# 원하는 변수
variable = '쓰레기통수'

# 범위 설정
vmin, vmax = gdf[variable].min(), gdf[variable].max()

# Plot 생성
fig, ax = plt.subplots(1, figsize=(10, 6))

# GeoDataFrame을 시각화
gdf.plot(column=variable, cmap='Greens', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 축 제거
ax.axis('off')

# 제목 추가
ax.set_title('쓰레기통 수 밀도', fontdict={'fontsize': '25', 'fontweight': '3'})

# 색상 바 추가
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm, ax=ax)

# 지도 저장
plt.savefig("map_no_background.png", dpi=300, bbox_inches='tight', pad_inches=0)

# 지도 출력
plt.show()
```
