```python
import pandas as pd
import folium
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
gdf = gpd.read_file('/content/drive/MyDrive/유성구 데이터/TL_SCCO_EMD.shp', encoding='euc-kr')
gdf
```

```python
df = pd.read_csv('/content/drive/MyDrive/유성구 데이터/최종데이터_정규화_클러스터.csv')
df
```

```python
# prompt: gdf의 EMD_KOR_NM과 df의 법정동명을 기준으로 df의 Cluster_K_means, Cluster_K_Medoids, Cluster_Hierarchical, Cluster_GMM 값을 추가

# Merge the dataframes based on 'EMD_KOR_NM' and '법정동명'
merged_gdf = gdf.merge(df, left_on='EMD_KOR_NM', right_on='법정동명', how='left')

# Display the merged GeoDataFrame
merged_gdf
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# 특정 법정동을 지정
target_dongs = ['봉명동', '관평동', '전민동', '장대동', '상대동', '지족동']

# 색상 매핑: 지정된 법정동은 #1b9461, 나머지는 #f2ffff
merged_gdf['color'] = merged_gdf['법정동명'].apply(lambda x: '#1b9461' if x in target_dongs else '#f2ffff')

# Plot 생성
fig, ax = plt.subplots(1, figsize=(10, 6))

# 색상 열을 사용하여 GeoDataFrame 시각화
merged_gdf.plot(color=merged_gdf['color'], linewidth=0.8, ax=ax, edgecolor='0.8', legend=False, alpha=1)

# 축 제거
ax.axis('off')

# 지도 저장
plt.savefig("5개선정_법정동.png", dpi=800, bbox_inches='tight', pad_inches=0)

# 지도 출력
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

variable = 'Cluster_Hierarchical'

# 각 클러스터 값에 대해 색상 매핑
color_map = {0: '#1b9461', 1: '#f2ffff', 2: '#b1e8d0'}

# 색상 열 추가 (Cluster_GMM의 값을 기반으로)
merged_gdf['color'] = merged_gdf[variable].map(color_map)

# Plot 생성
fig, ax = plt.subplots(1, figsize=(10, 6))

# 색상 열을 사용하여 GeoDataFrame 시각화
merged_gdf.plot(color=merged_gdf['color'], linewidth=0.8, ax=ax, edgecolor='0.8', legend=False, alpha=1)

# 축 제거
ax.axis('off')

# 지도 저장
plt.savefig("HC_cluster.png", dpi=800, bbox_inches='tight', pad_inches=0)

# 지도 출력
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

variable = 'Cluster_GMM'

# 범위 설정
vmin, vmax = merged_gdf[variable].min(), merged_gdf[variable].max()

# Plot 생성
fig, ax = plt.subplots(1, figsize=(10, 6))

# GeoDataFrame을 시각화, legend=False로 기본 범례 제거
color_map = ['#1b9461', '#76cba7', '#a7dcbe']
merged_gdf.plot(column=variable, color=color_map, linewidth=0.8, ax=ax, edgecolor='0.8', legend=False, alpha=1)

# 축 제거
ax.axis('off')

# 지도 저장
plt.savefig("map_no_background.png", dpi=800, bbox_inches='tight', pad_inches=0)

# 지도 출력
plt.show()
```
