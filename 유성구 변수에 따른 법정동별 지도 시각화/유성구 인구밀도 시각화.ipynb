```python
import geopandas as gpd
import folium
from folium import Choropleth, LayerControl
from shapely import wkt

# 데이터 준비 (이전 단계에서 이미 merged 데이터가 있다고 가정)
# DataFrame에서 GeoDataFrame으로 변환
# 'geometry' 컬럼이 WKT 문자열 형식으로 되어 있다고 가정
gdf = merged.copy()

# 'geometry' 컬럼을 shapely의 WKT 형태로 변환# GeoDataFrame으로 변환
gdf = gpd.GeoDataFrame(gdf, geometry='geometry')

# 1. CRS 설정 및 변환
gdf = gdf.set_crs(epsg=5179)  # 원본 데이터의 CRS 설정
gdf = gdf.to_crs(epsg=4326)   # EPSG:4326으로 변환

# 2. GeoDataFrame을 GeoJSON으로 변환
geojson_data = gdf.to_json()

# 3. 부산 중심부의 경도와 위도 (위도와 경도를 수정하여 원하는 중심으로 설정하세요)
center = [36.3668, 127.3451]  # 위도와 경도

# 4. 기본 지도 생성
m = folium.Map(location=center, zoom_start=10)

# 5. Choropleth 레이어 추가
choropleth = Choropleth(
    geo_data=geojson_data,
    data=gdf,
    columns=['행정동명', '인구밀도'],
    key_on='feature.properties.행정동명',  # GeoJSON 속성 확인 후 정확한 값 사용
    fill_color='YlOrRd',  # 색상 팔레트
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='인구밀도'
).add_to(m)

# 6. 레이어 컨트롤 추가
LayerControl().add_to(m)

# 7. 지도 저장
m.save('population_density_map.html')

# 지도 출력
m
```

```python
import geopandas as gpd
import folium

# 1. Shapefile 로드
gdf = gpd.read_file('/content/TL_SCCO_GEMD.shp')  # 실제 shapefile 경로로 수정하세요

# 2. CRS 설정 및 변환
gdf = gdf.set_crs(epsg=5179)  # 원본 CRS 설정
gdf = gdf.to_crs(epsg=4326)   # EPSG:4326으로 변환

# 3. GeoDataFrame을 GeoJSON으로 변환
geojson_data = gdf.to_json()

# 4. 부산 중심부의 경도와 위도 (위도와 경도를 수정하여 원하는 중심으로 설정하세요)
center = [36.3668, 127.3451]  # 위도와 경도

# 기본 지도 생성
m = folium.Map(location=center, zoom_start=10)

# GeoJSON 레이어 추가
folium.GeoJson(
    geojson_data,
    name='geojson'
).add_to(m)

# 레이어 컨트롤 추가
folium.LayerControl().add_to(m)

# 지도 저장
m.save('shapefile_map.html')

# 지도 출력
m
```
